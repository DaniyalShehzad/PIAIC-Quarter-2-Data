```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
```

```python
from tensorflow import keras
layers = keras.layers
models = keras.models
```

```python
# Step 1

df = pd.read_csv("datasets/sonar.csv")

df.head()
```

```python
# Step 2

df = df.sample(frac = 1)
df.head()
```

```python
# Step 3

len(df)
```

```python
training_data = df[:140]
training_data
```

```python
# Step 4

training_data_input = training_data.iloc[:,:-1].values
training_data_input
```

```python
training_data_labels = training_data.iloc[:,-1:].values
training_data_labels
```

```python
# Step 5

encoder = LabelEncoder()
encoder.fit(training_data_labels)

encoded_training = encoder.transform(training_data_labels)
encoded_training
```

```python
validation_data = df[140:160]
validation_data
```

```python
validation_data_input = validation_data.iloc[:,:-1].values
validation_data_input
```

```python
validation_data_labels = validation_data.iloc[:,-1:].values
validation_data_labels
```

```python
encoder.fit(validation_data_labels)

encoded_validation = encoder.transform(validation_data_labels)
encoded_validation
```

```python
testing_data = df[160:]
testing_data
```

```python
testing_data_input = testing_data.iloc[:,:-1].values
testing_data_input
```

```python
testing_data_labels = testing_data.iloc[:,-1:].values
testing_data_labels
```

```python
encoder.fit(testing_data_labels)

encoded_testing = encoder.transform(testing_data_labels)
encoded_testing
```

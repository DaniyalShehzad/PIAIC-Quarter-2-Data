```python
# Using HDF5 Format
#It is a file format
# stores data in binary form
# it is heirarchial data format
# we can process data directly from data in HDFS

# do at  home 3,15,16

import numpy as np
frame = pd.DataFrame({'a': np.random.randn(100)})

# hdfstore creates a file handler names "Store"
#hdfstore is a special function used to save hdf files
store = pd.HDFStore('mydata.h5')

#Stored dataframe in the store
store['obj1'] = frame                            # data is formed in the object

# we can store dataframe or a single column of the dataframe
store['obj1_col'] = frame['a']
print(store)

# to read the file back from store
x=store.obj1                        # x is the dataframe here stored in hdf format


print(type(x))
print(x.head())
print('--------------')
# putting the data in store and telling that the data will be in form of table
# if in table form, we can filter in the future
store.put('obj2', frame, format='table')

# to select data filterwise
# you can format data in index form
store.select('obj2', where=['index >= 10 and index <= 15'])
store.close()

#shortcut
#opens data processes and closes it
frame.to_hdf('mydata.h5', 'obj3', format='table')
frame = pd.read_hdf('mydata.h5', 'obj3', where=['index < 5'])
```

## Combining and Merging Datasets
* pandas.merge connects rows in DataFrames based on one or more keys.
* pandas.concat concatenates or stacks together objects along an axis
* The combine_first instance method enables splicing together overlapping data to fill in missing values in one object with values from another.


```python
import pandas as pd

df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})

df2 = pd.DataFrame({'key': ['a', 'b', 'd'],'data2': range(3)})


display(df1, "-------------------------")
display(df2, "-------------------------")
display(pd.merge(df1, df2, on='key'))

print("\n----------------------------------------------------------------------------------")
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
display(df3)
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
display(df4)

pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
df1 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'rkey': ['a', 'b', 'd'],'data2': range(3)})
pd.merge(df1, df2, left_on='lkey', right_on='rkey')
```

# concatenating Along an Axis
* Another kind of data combination operation is referred to interchangeably as concat enation, binding, or stacking

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
pd.concat([s1, s2, s3])
#By default concat works along axis=0
#pd.concat([s1, s2, s3], axis=1)
#In this case there is no overlap on the other axis, which as you can see is the sorted
#union (the 'outer' join) of the indexes. You can instead intersect them by passing
#join='inner' :
```

```python
s4 = pd.concat([s1, s3])
print(s4)
pd.concat([s1, s4], axis=1)
pd.concat([s1, s4], axis=1, join='inner')

#In this last example, the 'f' and 'g' labels disappeared because of the join='inner'
#option

#You can even specify the axes to be used on the other axes with join_axes :
#In [91]: pd.concat([s1, s4], axis=1, join_axes=[['a', 'c', 'b', 'e']])
```

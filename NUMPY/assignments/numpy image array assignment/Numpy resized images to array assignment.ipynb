```python
from PIL import Image
import glob
import numpy as np

image_list = []
resized_images = []


# reading images from folder
for filename in glob.glob(r'photos\*.jpg'):
    img = Image.open(filename)                                           #variable for images
    image_list.append(img)                                               # appending images to a list
        

# resizing images        
for image in image_list:                                                
    res = image.resize((200, 200))
    resized_images.append(res)                                           # appending resized images to another list 

    
mainarray = np.zeros([20, 200, 200, 3])                            # making the main array


i = 0
n = 0
for (i,new ) in enumerate(resized_images):
    img = np.array(new)
    mainarray[n] = img[:, :, :3]                                  # inserting image array in main array
    n = n + 1

    
    
print(mainarray)
```
